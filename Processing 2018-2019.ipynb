```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("season-1819_csv.csv")
```

```python
df.head()
```

```python
sum_column = df["FTHG"] + df["FTAG"]
df["Total goals scored in the match"] = sum_column
```

```python
sum_column = df["HS"] + df["AS"]
df["Total shots in the match"] = sum_column
```

```python
new_dataset = df.loc[:, ['Div','Date','HomeTeam','AwayTeam','FTHG','FTAG','Total goals scored in the match','FTR','HTHG','HTAG','HTR','Referee','HS','AS','Total shots in the match','HST','AST']]
```
